```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Price', 'Method', 'SellerG',
        'Distance', 'Bedroom2', 'Bathroom', 'Car',
       'Landsize', 'BuildingArea', 'CouncilArea',
        'Regionname', 'Propertycount']
dataset = dataset[cols_to_use]
```

Now I want to check if any of the columns have missing values.

```python
dataset.isna().sum()
```

Here now I'm going to fill the missing values...
Some of them are filled by zeros - this is done to the columns of those having very few missing values and to those features that can be zero.
Some of them are filled by the mean value(mean of all the available values in that particular column) -  this is done to those columns that have high number of missing values and also to those featrees that can't be zero i.e area etc.

Also we have to see what kind the feature is...

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
dataset.isna().sum()
```

```python
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace = True)
dataset.isna().sum()
```

Now I do one hot encoding for the text columns.

```python
dataset = pd.get_dummies(dataset, drop_first=True)
dataset.head()
```

```python
dataset.shape
```

Now I create X and y.
X- independent feature or the input label
y - target feature and here it is the price

```python
X = dataset.drop('Price', axis=1)
y = dataset.Price
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X,y,test_size=0.3,random_state=2)
# 30% testing and 70% data for training
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_X,train_y)
```

```python
reg.score(train_X,train_y), reg.score(test_X,test_y)
```

The score came out to be very low which is approximately 14% whereas the score for training dataset is 68% which clearly tell us that the model is overfitting. For training example the scoe wae fine but for test examples the score is horrible. In order to get rid over it I ,ake use of the regularisation i.e performing lasso and ridge regressions.

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha = 50, max_iter = 100, tol = 0.1)
lasso_reg.fit(train_X,train_y)
```

```python
lasso_reg.score(train_X,train_y), lasso_reg.score(test_X,test_y)
```

Here one can find that the accuracy of test data has been improved and is very close to that of train data. Hence overfitting has been improved.

```python
from sklearn.linear_model import Ridge


ridge_reg = Ridge(alpha=50, max_iter=100, tol=0.1) 
ridge_reg.fit(train_X, train_y)
```

```python
ridge_reg.score(train_X,train_y), ridge_reg.score(test_X, test_y) 
```

Here we can clearly see that the accuracy of the model with test data has further increased to 66.7%

So by using Lasso and Ridge regressions, the accuracy for the not seen datasamples(test data) improved a whole lot.
